# 2. MODELO PYTHON 

```python
from pyomo.environ import *
import pandas as pd

# 1. DEFININDO OS DADOS

def carregar_dados():

    CDs = ["CD1", "CD2", "CD3"]
    CCs = ["CC1", "CC2", "CC3", "CC4", "CC5"]

    custos_fixos = {
        "CD1": 0,
        "CD2": 500000,
        "CD3": 450000
    }

    oferta = {"CD1": 50, "CD2": 75, "CD3": 85}

    demandaA = {"CC1": 16, "CC2": 20, "CC3": 12, "CC4": 18, "CC5": 14}
    demandaB = {"CC1": 10, "CC2": 14, "CC3": 12, "CC4": 16, "CC5": 8}

    custoA = {
        ("CD1","CC1"):60, ("CD1","CC2"):65, ("CD1","CC3"):78, ("CD1","CC4"):67, ("CD1","CC5"):84,
        ("CD2","CC1"):45, ("CD2","CC2"):54, ("CD2","CC3"):76, ("CD2","CC4"):53, ("CD2","CC5"):32,
        ("CD3","CC1"):31, ("CD3","CC2"):43, ("CD3","CC3"):54, ("CD3","CC4"):65, ("CD3","CC5"):72
    }

    custoB = {
        ("CD1","CC1"):37, ("CD1","CC2"):39, ("CD1","CC3"):54, ("CD1","CC4"):45, ("CD1","CC5"):25,
        ("CD2","CC1"):23, ("CD2","CC2"):34, ("CD2","CC3"):21, ("CD2","CC4"):34, ("CD2","CC5"):70,
        ("CD3","CC1"):21, ("CD3","CC2"):38, ("CD3","CC3"):45, ("CD3","CC4"):38, ("CD3","CC5"):68
    }

    return CDs, CCs, custos_fixos, oferta, demandaA, demandaB, custoA, custoB


# 2. CRIANDO MODELO PYOMO

def montar_modelo():

    CDs, CCs, custos_fixos, oferta, demandaA, demandaB, custoA, custoB = carregar_dados()

    model = ConcreteModel("Problema_Localiza√ß√£o")

    # Conjuntos
    model.I = Set(initialize=CDs)
    model.J = Set(initialize=CCs)

    # Vari√°veis
    model.abre = Var(model.I, domain=Binary)
    model.envA = Var(model.I, model.J, domain=NonNegativeReals)
    model.envB = Var(model.I, model.J, domain=NonNegativeReals)

    # Fun√ß√£o Objetivo
    def objetivo(m):
        return (
            sum(custos_fixos[i] * m.abre[i] for i in m.I)
            + sum(custoA[i, j] * m.envA[i, j] for i in m.I for j in m.J)
            + sum(custoB[i, j] * m.envB[i, j] for i in m.I for j in m.J)
        )

    model.Z = Objective(rule=objetivo, sense=minimize)

    # Restri√ß√µes de capacidade
    def limiteA(m, i):
        return sum(m.envA[i, j] for j in m.J) <= oferta[i] * m.abre[i]

    def limiteB(m, i):
        return sum(m.envB[i, j] for j in m.J) <= oferta[i] * m.abre[i]

    model.cap_A = Constraint(model.I, rule=limiteA)
    model.cap_B = Constraint(model.I, rule=limiteB)

    # Restri√ß√µes de demanda
    def demanda_A(m, j):
        return sum(m.envA[i, j] for i in m.I) == demandaA[j]

    def demanda_B(m, j):
        return sum(m.envB[i, j] for i in m.I) == demandaB[j]

    model.demA = Constraint(model.J, rule=demanda_A)
    model.demB = Constraint(model.J, rule=demanda_B)

    return model


# 3. RESOLVENDO E FORMATANDO RESULTADOS

def resolver(model):

    solver = SolverFactory("highs")
    solver.solve(model)

    return model


# 4. GERANDO RELAT√ìRIOS

def gerar_relatorio(model):

    print("\n" + "="*55)
    print("üî∑ RESULTADOS DO PROBLEMA DE LOCALIZA√á√ÉO")
    print("="*55)
    print(f"‚û° Custo m√≠nimo total: R$ {model.Z():,.2f}")
    print("-"*55)

    # CDs instalados
    print("\nüìå Centros instalados:")
    for i in model.I:
        status = "INSTALADO" if model.abre[i]() == 1 else "FECHADO"
        print(f" ‚Ä¢ {i}: {status}")

    # Fluxo A
    print("\nüöö Fluxo de transporte do produto A:")
    dadosA = []
    for i in model.I:
        for j in model.J:
            v = model.envA[i,j]()
            if v > 0:
                dadosA.append([i, j, v])

    dfA = pd.DataFrame(dadosA, columns=["CD", "CC", "Quantidade"])
    print(dfA.to_string(index=False))

    # Fluxo B
    print("\nüöõ Fluxo de transporte do produto B:")
    dadosB = []
    for i in model.I:
        for j in model.J:
            v = model.envB[i,j]()
            if v > 0:
                dadosB.append([i, j, v])

    dfB = pd.DataFrame(dadosB, columns=["CD", "CC", "Quantidade"])
    print(dfB.to_string(index=False))

    print("\n" + "="*55)
    print("Fim da Solu√ß√£o")
    print("="*55)


# EXECUTANDO

if __name__ == "__main__":
    modelo = montar_modelo()
    modelo = resolver(modelo)
    gerar_relatorio(modelo)
```
